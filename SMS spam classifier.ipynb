```python
import pandas as pd
import kagglehub
import  os
# Download latest version
path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")


file_path = os.path.join(path,"spam.csv")
df=pd.read_csv(file_path, encoding='latin-1')

```

```python

df.drop(columns=["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], inplace=True, errors='ignore')
df.columns=['labels','text']
df.head(50)
```

```python
#Here, we will analyse the number of spams, and hams present in the dataset, and plot them.
import matplotlib.pyplot as plt
counts=df['labels'].value_counts()
print("Total number of SMS")
print(counts)

plt.figure(figsize=(7,7))
plt.pie(counts, labels=['Ham','Spam'],autopct='%1.1f%%',startangle=90,explode=(0,0.03),shadow=True)
plt.title("Comparison between the frequency of spam and genuine messages.",fontsize=14)
plt.axis('equal')
plt.show()
```

```python
df.drop_duplicates(inplace=True)
```

```python
import string

df['char_count']=df['text'].apply(len)
df['word_count']=df['text'].apply(lambda x: len(x.split()))
df['sent_count']=df['text'].apply(lambda x: len(x.split('.')))

df.head(50)
```

```python
df[df['labels']=='ham'][['word_count','sent_count','char_count']].describe()
```

```python
df[df['labels']=='spam'][['word_count','sent_count','char_count']].describe()
```

```python
# This calculates the mean (average) for every numerical column, grouped by Spam/Ham
averages = df.groupby('labels')[['char_count', 'word_count', 'sent_count']].mean()
print(averages)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
custom_colors=("Yellow","black")
sns.histplot(data=df,x=df['word_count'],hue='labels',palette=custom_colors,fill=True,bins=150)
plt.title("Comparison between the word counts of spam and ham")
plt.show()
```

```python
plt.figure(figsize=(10,6))
colors=("Yellow","Black")
sns.histplot(data=df,x=df['char_count'],hue='labels',palette=colors,fill=True,bins=150)
plt.title("Comparison between the number of characters used in Ham and Spam")
plt.show()
```
